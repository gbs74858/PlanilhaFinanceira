```python
# ==========================================================
# PROJETO: Calculadora e An√°lise de Porcentagens com Python
# Autor: Gabriel Santos
# Ambiente: Google Colab (compat√≠vel com celular)
# ==========================================================
from IPython.display import Markdown

Markdown("""
# üßÆ Calculadora e Analisador de Porcentagens
Este projeto demonstra como funcionam os c√°lculos de porcentagem, com:

- Calculadora interativa
- Exerc√≠cios autom√°ticos
- Gr√°ficos com dados percentuais
- Explica√ß√µes passo a passo

Ideal para estudos, projetos escolares ou postagem no GitHub!
""")
def porcentagem(valor, porc):
    return valor * (porc / 100)

def parte_para_total(parte, porc):
    return (parte * 100) / porc

def porcentagem_relacao(valor1, valor2):
    return (valor2 / valor1) * 100
import random

def menu():
    print("\n==============================")
    print("  CALCULADORA DE PORCENTAGEM")
    print("==============================")
    print("1 - Calcular porcentagem")
    print("2 - Achar o valor total a partir de porcentagem")
    print("3 - Saber quantos % um valor representa de outro")
    print("4 - Gerar exerc√≠cios")
    print("5 - Visualizar gr√°fico")
    print("6 - Sair")

def executar_menu():
    while True:
        menu()
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            valor = float(input("Valor: "))
            porc = float(input("Porcentagem: "))
            print("Resultado:", porcentagem(valor, porc))

        elif opcao == "2":
            parte = float(input("Parte: "))
            porc = float(input("Porcentagem: "))
            print("Total:", parte_para_total(parte, porc))

        elif opcao == "3":
            v1 = float(input("Valor 1: "))
            v2 = float(input("Valor 2: "))
            print("Percentual:", porcentagem_relacao(v1, v2))

        elif opcao == "4":
            gerar_exercicios()

        elif opcao == "5":
            gerar_grafico()

        elif opcao == "6":
            print("Saindo...")
            break

        else:
            print("Op√ß√£o inv√°lida!")
def gerar_exercicios():
    acertos = 0
    print("\nGerando 5 exerc√≠cios...\n")

    for i in range(5):
        valor = random.randint(100, 500)
        porc = random.randint(5, 40)
        correto = porcentagem(valor, porc)

        print(f"{i+1}) Quanto √© {porc}% de {valor}?")
        resp = float(input("Sua resposta: "))

        if abs(resp - correto) < 0.1:
            print("‚úî Correto!\n")
            acertos += 1
        else:
            print(f"‚úò Errado! Resposta correta: {correto}\n")

    print(f"Voc√™ acertou {acertos}/5")
import matplotlib.pyplot as plt

def gerar_grafico():
    categorias = ["A", "B", "C", "D", "E"]
    valores = [random.randint(10, 90) for _ in categorias]

    plt.figure(figsize=(6,4))
    plt.bar(categorias, valores)
    plt.title("Distribui√ß√£o Percentual Aleat√≥ria")
    plt.xlabel("Categoria")
    plt.ylabel("Porcentagem (%)")
    plt.show()
executar_menu()
```
