```python
# =============================================
# PROJETO COMPLETO: Analisador Geométrico 2D
# =============================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print("=== ANALISADOR GEOMÉTRICO ===\n")

# Entrada dos pontos
x1 = float(input("Digite x1: "))
y1 = float(input("Digite y1: "))

x2 = float(input("Digite x2: "))
y2 = float(input("Digite y2: "))

print("\nPonto A =", (x1, y1))
print("Ponto B =", (x2, y2))

# ------------------------------------
# Cálculo da distância
# ------------------------------------
distancia = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
print(f"\nDistância entre A e B: {distancia:.4f}")

# ------------------------------------
# Coeficiente angular e linear
# ------------------------------------
if x1 == x2:
    print("\nA reta é vertical.")
    print("Equação: x =", x1)
    vertical = True
else:
    vertical = False
    m = (y2 - y1) / (x2 - x1)     # coeficiente angular
    b = y1 - m * x1                # coeficiente linear
    print(f"\nCoeficiente angular m = {m:.4f}")
    print(f"Coeficiente linear b = {b:.4f}")
    print(f"Equação da reta: y = {m:.4f}x + {b:.4f}")

    # Análise de crescimento
    if m > 0:
        print("A reta é CRESCENTE.")
    elif m < 0:
        print("A reta é DECRESCENTE.")
    else:
        print("A reta é CONSTANTE.")

# ------------------------------------
# Criar tabela de valores
# ------------------------------------
if not vertical:
    x_vals = np.linspace(min(x1, x2) - 5, max(x1, x2) + 5, 11)
    y_vals = m * x_vals + b
    tabela = pd.DataFrame({"x": x_vals, "y": y_vals})
    print("\nTabela de valores:")
    tabela
else:
    x_vals = np.array([x1] * 11)
    y_vals = np.linspace(min(y1, y2) - 5, max(y1, y2) + 5, 11)
    tabela = pd.DataFrame({"x": x_vals, "y": y_vals})
    print("\nTabela de valores (reta vertical):")
    tabela

# ------------------------------------
# Gráfico
# ------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Reta AB", linewidth=3)
plt.scatter([x1, x2], [y1, y2], color="red", s=100, label="Pontos A e B")

plt.title("Representação Gráfica da Reta")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.grid(True)
plt.legend()
plt.show()

# ------------------------------------
# Comparação com terceira reta (opcional)
# ------------------------------------
print("\n== Comparação com uma terceira reta ==")
print("Forma: y = mx + b")

resposta = input("Deseja comparar com outra reta? (s/n): ").lower()

if resposta == "s":
    m2 = float(input("Digite m2: "))
    b2 = float(input("Digite b2: "))

    print(f"\nReta 1: y = {m:.4f}x + {b:.4f}")
    print(f"Reta 2: y = {m2:.4f}x + {b2:.4f}\n")

    if m == m2:
        if b == b2:
            print("As retas são COINCIDENTES.")
        else:
            print("As retas são PARALELAS.")
    else:
        x_int = (b2 - b) / (m - m2)
        y_int = m * x_int + b
        print(f"As retas são SECANTES e se interceptam em ({x_int:.4f}, {y_int:.4f}).")
```
