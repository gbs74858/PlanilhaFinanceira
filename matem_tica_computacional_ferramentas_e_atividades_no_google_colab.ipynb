```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print("Função do 1º Grau: f(x) = ax + b\n")

a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))

# Raiz
if a != 0:
    raiz = -b / a
    print(f"\nRaiz da função: x = {raiz}")
else:
    print("\nA função é constante")

# Gerar tabela
x = np.arange(-10, 11)
y = a * x + b
df = pd.DataFrame({"x": x, "f(x)": y})
df

# Gráfico
plt.plot(x, y, linewidth=3)
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title("Função do 1º Grau")
plt.grid(True)
plt.show()
import numpy as np
import matplotlib.pyplot as plt

print("Função Quadrática: f(x) = ax² + bx + c\n")

a = float(input("Digite a: "))
b = float(input("Digite b: "))
c = float(input("Digite c: "))

# Delta
delta = b**2 - 4*a*c
print(f"\n∆ = {delta}")

# Vértice
xv = -b/(2*a)
yv = -delta/(4*a)
print(f"Vértice: ({xv}, {yv})")

# Raízes
if delta > 0:
    r1 = (-b + delta**0.5)/(2*a)
    r2 = (-b - delta**0.5)/(2*a)
    print(f"Raízes reais: {r1} e {r2}")
elif delta == 0:
    r = -b/(2*a)
    print(f"Raiz dupla: {r}")
else:
    print("Não possui raízes reais.")

# Gráfico
x = np.linspace(xv - 10, xv + 10, 400)
y = a*x**2 + b*x + c

plt.plot(x, y, linewidth=3)
plt.scatter([xv], [yv], color="red")
plt.title("Função Quadrática")
plt.grid(True)
plt.show()
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

print("Cálculo Estatístico\n")

valores = input("Digite valores separados por espaço: ").split()
valores = [float(x) for x in valores]

media = np.mean(valores)
mediana = np.median(valores)
contagem = Counter(valores)
moda = contagem.most_common(1)[0][0]

print(f"\nMédia: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

# Gráfico
plt.bar(range(len(valores)), valores)
plt.title("Valores informados")
plt.grid(True)
plt.show()
import numpy as np

print("Sistema 2x2\n")

a = float(input("Digite a: "))
b = float(input("Digite b: "))
c = float(input("Digite c: "))

d = float(input("Digite d: "))
e = float(input("Digite e: "))
f = float(input("Digite f: "))

A = np.array([[a, b], [d, e]])
B = np.array([c, f])

solucao = np.linalg.solve(A, B)

print(f"\nSolução: x = {solucao[0]}, y = {solucao[1]}")
import random

print("Gerador de Exercícios de Matemática\n")

for i in range(10):
    a = random.randint(1, 20)
    b = random.randint(1, 20)
    op = random.choice(["+", "-", "*", "/", "**"])

    if op == "/":
        b = random.randint(1, 10)

    print(f"{i+1}) {a} {op} {b} = ?")
    print(f"Resposta: {eval(str(a)+op+str(b))}\n")
```
